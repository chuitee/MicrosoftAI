```python
aveMonthSpend= pd.read_csv('AW_AveMonthSpend.csv')
```

```python
def clean_month_spend_data(aveMonthSpend):
    'Function to load the aveMonthSpend data set from a .csv file' 
    import pandas as pd
    import numpy as np
    
    ## Recode names
    ## fix column names so the '-' character becomes '_'
    cols = aveMonthSpend.columns
    aveMonthSpend.columns = [str.replace('-', '_') for str in cols]
    
    ## Treat missing values
    ## Remove rows with missing values, accounting for mising values coded as '?'
    ## cols = ['CustomerID', 'AveMonthSpend']
    ## for column in cols:
    ##     aveMonthSpend.loc[aveMonthSpend[column] == '?', column] = np.nan
    ## aveMonthSpend.dropna(axis = 0, inplace = True)

    ## Transform column data type
    ## Convert some columns to numeric values
    for column in cols:
        aveMonthSpend[column] = pd.to_numeric(aveMonthSpend[column])

    return aveMonthSpend
aveMonthSpend = clean_month_spend_data(aveMonthSpend)

print(aveMonthSpend.columns)
```
