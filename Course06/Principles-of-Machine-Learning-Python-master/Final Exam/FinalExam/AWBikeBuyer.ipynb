```python
bike_buyer = pd.read_csv('AW_BikeBuyer.csv')
```

```python
def clean_bike_buyer_data(bike_buyer):
    'Function to load the bike_buyer data set from a .csv file' 
    import pandas as pd
    import numpy as np
    
    ## Recode names
    ## fix column names so the '-' character becomes '_'
    cols = bike_buyer.columns
    bike_buyer.columns = [str.replace('-', '_') for str in cols]
    
    ## Treat missing values
    ## Remove rows with missing values, accounting for mising values coded as '?'
    ## cols = ['CustomerID', 'AveMonthSpend']
    ## for column in cols:
    ##     aveMonthSpend.loc[aveMonthSpend[column] == '?', column] = np.nan
    ## aveMonthSpend.dropna(axis = 0, inplace = True)

    ## Transform column data type
    ## Convert some columns to numeric values
    for column in cols:
        bike_buyer[column] = pd.to_numeric(bike_buyer[column])

    return bike_buyer
bike_buyer = clean_bike_buyer_data(bike_buyer)

print(bike_buyer.columns)
```
